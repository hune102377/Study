#### 정규 표현식 예제 #1

```python
import re

# compile() 사용 안함
# 정규식 객체 생성 안함
m = re.match('[a-z]+', 'Python') # 정확히 소문자로 구성되어있나 검사 => None
print(m)
print(re.search('apple', 'I like apple!'))

# compile() 사용 
# 정규식 객체 p 생성
p = re.compile('[a-z]+')
m = p.match('python')
print(m)
print(re.search('apple', 'I like apple 123'))
```

```python
# findall() 함수 : 일치하는 모든 문자열을 리스트로 리턴

p = re.compile('[a-z]+')
print(p.findall('life is too short'))
```

```python
# search() 함수
result = p.search('I like apple 123') # 처음 매칭되는 문자열 리턴
print(result)


result = p.findall('I like apple 123') # pattern과 매치되는 모든 경우를 찾아 리스트로 반환
print(result)
```

### Match 메소드 예제
#### - 전화 번호 분석

```python
import re

# ^ . $ 을 명시해야 정확한 자리수 검사가 이루어짐
tel_checker = re.compile("^(\d{2,})-(\d{3,4})-(\d{4})$")

print(tel_checker.match('02-123-4567'))
print(tel_checker.match('053-950-45678'))
print(tel_checker.match('053123-4567'))
```

```python
# group() 매칭된 전체 문자열 반환 
# group(index) : 해당 인덱스에 매칭된 문자열 반환, index = 0 : 전체리턴

m = tel_checker.match('02-123-4567')

print('group(): ', m.group())
print('group(1): ', m.group(1))
print('group(2): ', m.group(2,3))
print('strat(): ', m.start())
print('end(): ', m.end())
```

### 전방 긍정과 전방 부정 탐색

```python
import re

# 전방 긍정 탐색 (. :  임의의 문자)
lookahead1 = re.search('. + (?=won)', '1000 won')
print(lookahead1)

lookahead2 = re.search('. + (?=log)', '2022-07-01 00:00:01 ABC.log: 전방탐색')
print(lookahead2)

# 전방 부정 탐색
lookahead3 = re.search('\d{4}(?!-)', '010-1234-5678') # 4개의 숫자 다음에 -가 없는 경우
print(lookahead3)
```
