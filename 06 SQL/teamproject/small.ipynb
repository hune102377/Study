```python
font_path = r'c:\Windows\Fonts\malgun.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
df_target = pd.read_csv('./Data/target.csv', encoding='cp949', header = 1)
df_target.drop(df_target.columns[0], axis=1, inplace = True)
df_target.drop(df_target.columns[0], axis=1, inplace = True)

list_text = []
list_num = []

for i in range(0,13):
    list_text.append(df_target.iloc[0][i])
    list_num.append(df_target.iloc[1][i])

fig = plt.figure(figsize=(10,8)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 배경색을 하얀색으로 설정
ax = fig.add_subplot() ## 프레임 생성
 
pie = ax.pie(list_num, ## 파이차트 출력
       startangle=90, ## 시작점을 90도(degree)로 지정
       counterclock=False, ## 시계 방향으로 그린다.
       autopct=lambda p : '{:.2f}%'.format(p), ## 퍼센티지 출력
       wedgeprops=dict(width=0.5), ## 중간의 반지름 0.5만큼 구멍을 뚫어준다.
       labels = list_text
       )
 
plt.legend(pie[0], list_text[:4]) ## 범례 표시
plt.title('경제적 도움이 필요한 집단', fontsize = 14)
plt.show()
```
