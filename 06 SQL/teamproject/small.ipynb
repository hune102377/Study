```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import numpy as np
import platform
```

```python
font_path = r'c:\Windows\Fonts\malgun.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
df_target = pd.read_csv('./Data/target.csv', encoding='cp949', header = 1)
df_target.drop(df_target.columns[0], axis=1, inplace = True)
df_target.drop(df_target.columns[0], axis=1, inplace = True)

list_text = []
list_num = []

for i in range(0,13):
    list_text.append(df_target.iloc[0][i])
    list_num.append(df_target.iloc[1][i])

fig = plt.figure(figsize=(10,8)) ##  
fig.set_facecolor('white') ##    
ax = fig.add_subplot() ##  
 
pie = ax.pie(list_num, ##  
       startangle=90, ##  90(degree) 
       counterclock=False, ##   .
       autopct=lambda p : '{:.2f}%'.format(p), ##  
       wedgeprops=dict(width=0.5), ##   0.5  .
       labels = list_text
       )
 
plt.legend(pie[0], list_text[:4]) ##  
plt.title('   ', fontsize = 14)
plt.show()
```
